```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_csv("polynomial.csv",sep=";")
df.head()
```

```python
polynomial_regression = PolynomialFeatures(degree=2)
```

```python
#veri setinin grafiğini çizdirelim
plt.scatter(df["deneyim"],df["maas"])
plt.xlabel("Deneyim (Yıl)")
plt.ylabel("Maaş")
plt.savefig("1.png")
plt.show()
```

```python
#doğru olmayan lineer regression çizimini yapıp hataları görelim
reg = LinearRegression()
reg.fit(df[["deneyim"]],df["maas"])

plt.xlabel("Deneyim (Yıl)")
plt.ylabel("Maaş")
plt.scatter(df["deneyim"],df["maas"])

x = df["deneyim"]
y = reg.predict(df[["deneyim"]])
plt.plot(x,y,color="green",label = "LinearRegression")
plt.legend()
plt.show()
```

```python
polynomial_regression = PolynomialFeatures(degree=4)
x_polynomial = polynomial_regression.fit_transform(df[["deneyim"]])
```

```python
#modelimizi eğitelim
reg = LinearRegression()
reg.fit(x_polynomial,df["maas"])
```

```python
y_head = reg.predict(x_polynomial)
plt.plot(df["deneyim"],y_head,color="red",label="Polynomial Regression")
plt.legend()
plt.scatter(df["deneyim"],df["maas"])
plt.show()
```

```python
#şimdi yeni bir deneyim level oluşturup maaşın ne kadar olabileceğine dair bir tahmin oluşturalım
x_polynomial_tahmin = polynomial_regression.fit_transform([[4.5]])
reg.predict(x_polynomial_tahmin)
```
